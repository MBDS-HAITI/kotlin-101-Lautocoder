```kotlin
for (i in 1..5) print(i)
println()
for (i in 5 downTo 1) print(i)
println()
for (i in 3..6 step 2) print(i)
println()
for (i in 'd'..'g') print (i)
println()
for (i in 'g' downTo 'd' step 2) print (i)
```

```kotlin
fun encodeMsg(msg: String, encode: (String) -> String): String {
    return encode(msg)
}
val enc1: (String) -> String = { input -> input.reversed() }
fun enc2(input:String): String = input.reversed()

println(encodeMsg(encodeMsg("abc", enc1), enc1))
println(encodeMsg(encodeMsg("abc", { input -> input.reversed() }), { input -> input.reversed() }))
```
